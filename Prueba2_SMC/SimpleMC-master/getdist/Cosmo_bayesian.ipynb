# Notebook para el artículo cosmología observacional.
### En este notebook se elaborarán las gráficas para el artículo de cosmología observacional.
### El modelo LCDM es el modelo cosmológico estándar.
### El modelo cLCDM es $\Lamda$ CDM donde el parámetro de densidad asociado a la curvatura se ha dejado libre para que sea ajustado por el código SimpleMc.
#### Está basado en el notebook dr12_constraints.

```python
%matplotlib inline
import matplotlib.pyplot as plt
from getdist import plots
import planckStyle
import pylab
import numpy as np

#dir_name = '/Users/josevazquezgonzalez/Desktop/BOSS/BOSS_files/DR12/chains'
dir_name = '/home/osbuster/Documentos/Prueba2_SMC/SimpleMC-master/chains'

params1 = {'backend': 'pdf',
               'axes.labelsize': 20,
               'xtick.labelsize': 20,
               'ytick.labelsize': 20,
               'legend.fontsize': 16,
               'lines.markersize': 6,
               'font.size': 25,
               'text.usetex': True}#
pylab.rcParams.update(params1)
```

# $\Lambda$DCM.

```python
#g = planckStyle.getSinglePlotter(chain_dir = dir_name)
roots = ['LCDM_phy_BBAO', 'LCDM_phy_BBAO+Planck', 'LCDM_phy_BBAO+Planck+SN']

g= plots.getSinglePlotter(chain_dir = dir_name)

#Varias graficas.
g.plot_2d(roots, param_pair=['Obh2','Om'], filled=True)
g.add_legend(['BBAO', 'BBAO+Planck', 'BBAO+Planck+SN'], colored_text=True,  legend_loc='upper right')
g.export('Obh2_Om_LCDM_todas.pdf')
```

```python
#g = planckStyle.getSinglePlotter(chain_dir = dir_name)
roots = ['LCDM_phy_BBAO', 'LCDM_phy_BBAO+Planck', 'LCDM_phy_BBAO+Planck+SN']

g= plots.getSinglePlotter(chain_dir = dir_name)

#Varias graficas.
g.plot_2d(roots, param_pair=['h','Om'], filled=True)
g.add_legend(['BBAO', 'BBAO+Planck', 'BBAO+Planck+SN'], colored_text=True,  legend_loc='upper right')
g.export('h_Om_LCDM_todas.pdf')
```
